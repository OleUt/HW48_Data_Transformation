```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns
import statsmodels.api as sm

from sklearn.preprocessing import PowerTransformer
```

```python
df=pd.read_csv("Data_to_Transform.csv")  
df
```

```python
####### 1. Moderate Positive Skew

sns.histplot(df['Moderate Positive Skew'])
plt.title('histogram')

fig = plt.figure()
fig = sm.qqplot(df['Moderate Positive Skew'], line='45')
plt.title('qq plot')
plt.show()
```

```python
# перетворення квадратним коренем

df_sqrt = np.sqrt(df['Moderate Positive Skew'])
sns.histplot(df_sqrt)
plt.title('histogram')

fig = plt.figure()
fig = sm.qqplot(df_sqrt, line='45')
plt.title('qq plot')
plt.show()
```

```python
# перетворення кубічним коренем

df_cbrt = np.cbrt(df['Moderate Positive Skew'])
sns.histplot(df_cbrt)
plt.title('histogram')

fig = plt.figure()
fig = sm.qqplot(df_cbrt, line='45')
plt.title('qq plot')
plt.show()
```

```python
# логарифмічне перетворення

df_log = np.log(df['Moderate Positive Skew'])
sns.histplot(df_log)
plt.title('histogram')

fig = plt.figure()
fig = sm.qqplot(df_log, line='45')
plt.title('qq plot')
plt.show()
```

```python
# перетворення Бокса-Кокса

data = df['Moderate Positive Skew'].values.reshape(-1, 1)
pt = PowerTransformer(method='box-cox')
df_boxcox = pt.fit_transform(data)
sns.histplot(df_boxcox)
plt.title('histogram')

fig = plt.figure()
fig = sm.qqplot(df_boxcox, line='45')
plt.title('qq plot')
plt.show()
```

```python
##### 3. Moderate Negative Skew

sns.histplot(df['Moderate Negative Skew'])
plt.title('histogram')

fig = plt.figure()
fig = sm.qqplot(df['Moderate Negative Skew'], line='45')
plt.title('qq plot')
plt.show()
```

```python
# Transforming Negatively Skewed Data with the Square Root Method

df_sqrt = np.sqrt(max(df['Moderate Negative Skew']+1) - df['Moderate Negative Skew'])
sns.histplot(df_sqrt)
plt.title('histogram')

fig = plt.figure()
fig = sm.qqplot(df_sqrt, line='45')
plt.title('qq plot')
plt.show()
```

```python
# Transforming Negatively Skewed Data with the Cube Root Method

df_cbrt = np.cbrt(max(df['Moderate Negative Skew']+1) - df['Moderate Negative Skew'])
sns.histplot(df_cbrt)
plt.title('histogram')

fig = plt.figure()
fig = sm.qqplot(df_cbrt, line='45')
plt.title('qq plot')
plt.show()
```

```python
# Log Transforming Negatively Skewed Data

df_log = np.log(max(df['Moderate Negative Skew'] + 1) - df['Moderate Negative Skew'])
sns.histplot(df_log)
plt.title('histogram')

fig = plt.figure()
fig = sm.qqplot(df_log, line='45')
plt.title('qq plot')
plt.show()
```

```python
# Transforming Negatively Skewed Data with the Box-Cox method

data = (max(df['Moderate Negative Skew'] + 1)- df['Moderate Negative Skew']).values.reshape(-1, 1)
pt = PowerTransformer(method='box-cox')
df_boxcox = pt.fit_transform(data)
sns.histplot(df_boxcox)
plt.title('histogram')

fig = plt.figure()
fig = sm.qqplot(df_boxcox, line='45')
plt.title('qq plot')
plt.show()
```
